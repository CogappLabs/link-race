# Visualisation TSV

```python
import pandas as pd
```

```python
with open('./visualisation_data_with_clusters.tsv') as myfile:
    mydata = (line.replace('\n', '').split('\t') for line in myfile)
    header = next(mydata)
    df = pd.DataFrame(mydata, columns=header)
```

```python
df.head()
```

```python
def get_coords(df, label):
    return df[df['label'].str.lower().str.startswith(label, na=False)][['label', 'x', 'y']].sort_values(by=['label'])
```

```python
get_coords(df, 'joy division')
```
