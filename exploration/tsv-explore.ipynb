# Visualisation TSV

```python
import pandas as pd
```

```python
with open('./visualisation_data_with_clusters.tsv') as myfile:
    mydata = (line.replace('\n', '').split('\t') for line in myfile)
    header = next(mydata)
    df = pd.DataFrame(mydata, columns=header)
```

```python
df.head()
```

```python
def get_coords(df, label, sort=False):
    out = df[df['label'].str.lower().str.startswith(label, na=False)][['label', 'id', 'x', 'y']]
    if sort:
        return out.sort_values(by=['label'])
    else:
        return out
```

```python
get_coords(df, 'new order', sort=True)
```

```python
get_coords(df, 'queen victoria', sort=True)
```
